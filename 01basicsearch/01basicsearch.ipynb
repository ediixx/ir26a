```python
# Lista los archivos de la carpeta y carga el contenido de cada uno en un diccionario.
import os

path = "C:\\Users\\wwwed\\OneDrive - Escuela Politécnica Nacional\\Escritorio\\ir26a\\data\\gutenberg\\data"
os.listdir(path)

contenidos = {}
for archivo in os.listdir(path):
    with open(os.path.join(path, archivo), 'r', encoding='utf-8') as f:
        contenidos[archivo] = f.read()
        print (contenidos[archivo][:100])
```

```python
# Abre varios libros específicos y guarda su contenido en un diccionario.
path = "C:\\Users\\wwwed\\OneDrive - Escuela Politécnica Nacional\\Escritorio\\ir26a\\data\\gutenberg\\data"
files = [path + "\\pg84.txt", path + "\\pg100.txt", path + "\\pg2701.txt"]
contenidos = {}

for file in files:
    with open(file, "r", encoding="utf-8") as f:
        contenido = f.read()
        contenidos[file] = contenido
        print(contenido [:100])
```

```python
# Busca una palabra en los textos cargados y mide el tiempo de ejecución.
import time

start_time = time.time()
query = 'Mary'
resultados = []

for key in contenidos.keys():
    if query in contenidos[key]:
        resultados.append(key)
        

print (f"Los archivos que contienen la palabra {query} son: {resultados}")
end_time = time.time()
print(f"Tiempo de ejecución: {end_time - start_time} segundos")
```

```python
# Define una función que devuelve los archivos donde aparece una palabra.
def buscar(query, contenidos):
    resultados = []
    for key in contenidos.keys():
        if query in contenidos[key]:
            resultados.append(key)
    return resultados
```

```python
# Recorre todas las palabras únicas y aplica la búsqueda para medir el tiempo total.
import time

start_time = time.time()
for p in palabras:
    buscar(p, contenidos)
end_time = time.time()
print(f"Tiempo de ejecución: {end_time - start_time} segundos") 
```
